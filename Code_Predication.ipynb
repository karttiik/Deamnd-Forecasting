```
from pyspark.sql import SQLContext
fileName="dbfs:/FileStore/tables/Cement_Company_Data_Model.csv"
data= spark.read.csv(fileName,inferSchema = "true",header="true")
data.show()
```

```
from pyspark.sql import functions as F
df = data.withColumn("Difference",F.col("Target (MT)") - F.col("Production (MT)"))
df.show()
```

```
sum1 = 0
sum2 = 0
sum3 = 0
sum4 = 0
sum5 = 0
sum6 = 0

sum_diff1 = 0
sum_diff2 = 0
sum_diff3 = 0
sum_diff4 = 0
sum_diff5 = 0
sum_diff6 = 0

data_itr = df.rdd.toLocalIterator()

# looping thorough each row of the dataframe
for row in data_itr:
	
	# while looping through each row printing
	# the data of Id, Job Profile and City
    if (row['PlantId'] == 1):
      sum1 = sum1 + row["Production (MT)"]
      sum_diff1 = sum_diff1 + row["Difference"]
    if (row['PlantId'] == 2):
      sum2 = sum2 + row["Production (MT)"]
      sum_diff2 = sum_diff2 + row["Difference"]
    if (row['PlantId'] == 3):
      sum3 = sum3 + row["Production (MT)"]
      sum_diff3 = sum_diff3 + row["Difference"]
    if (row['PlantId'] == 4):
      sum4 = sum4 + row["Production (MT)"]
      sum_diff4 = sum_diff4 + row["Difference"]
    if (row['PlantId'] == 5):
      sum5 = sum5 + row["Production (MT)"]
      sum_diff5 = sum_diff5 + row["Difference"]
    if (row['PlantId'] == 6):
      sum6 = sum6 + row["Production (MT)"]
      sum_diff6 = sum_diff6 + row["Difference"]

      
#Production value for each plant in 2021
#no. of years = 6
Production2021_1 = sum1/6
Production2021_2 = sum2/6
Production2021_3 = sum3/6
Production2021_4 = sum4/6
Production2021_5 = sum5/6
Production2021_6 = sum6/6

#Difference value for each plant in 2021
Diff2021_1 = sum_diff1/6
Diff2021_2 = sum_diff2/6
Diff2021_3 = sum_diff3/6
Diff2021_4 = sum_diff4/6
Diff2021_5 = sum_diff5/6
Diff2021_6 = sum_diff6/6

print(Production2021_1)
print(Production2021_2)
print(Production2021_3)
print(Production2021_4)
print(Production2021_5)
print(Production2021_6)
```

```
#Predicting Target value using formula AvgProduction - AvgDiff
Target2021_1 = Production2021_1 - Diff2021_1
Target2021_2 = Production2021_2 - Diff2021_2
Target2021_3 = Production2021_3 - Diff2021_3
Target2021_4 = Production2021_4 - Diff2021_4
Target2021_5 = Production2021_5 - Diff2021_5
Target2021_6 = Production2021_6 - Diff2021_6

print(Target2021_1)
print(Target2021_2)
print(Target2021_3)
print(Target2021_4)
print(Target2021_5)
print(Target2021_6)
```
